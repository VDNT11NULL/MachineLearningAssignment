Let us see in detail the functioning of MinMaxScaler.

### Min-Max Scaling

Min-Max Scaling, also known as normalization, is a technique used to transform features to a common scale.
 This is particularly useful when features have different ranges and units, and it helps algorithms that are sensitive to feature scales.

#### Functioning

Min-Max Scaling transforms each feature to a fixed range, typically [0, 1]. The formula used for Min-Max scaling is:

$$ x_{\text{scaled}} = \frac{x - \text{min}(x)}{\text{max}(x) - \text{min}(x)} $$

where:
- $ x $ is the original feature value.
- $ \text{min}(x) $ is the minimum value of the feature.
- $ \text{max}(x) $ is the maximum value of the feature.
- $ x_{\text{scaled}} $ is the scaled feature value.
